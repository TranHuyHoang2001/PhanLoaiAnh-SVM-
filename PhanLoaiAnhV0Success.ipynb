os: là thư viện hỗ trợ với các tác vụ hệ thống như xử lý thư mục, tạo, xóa, đổi tên file, ...

matplotlib.pyplot: là thư viện hỗ trợ với việc vẽ đồ thị

skimage.io.imread: là hàm trong thư viện skimage, hỗ trợ đọc hình ảnh từ file

skimage.transform.resize: là hàm trong thư viện skimage, hỗ trợ thay đổi kích cỡ của hình ảnh.

```python
import os 
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target = []
images = []
flat_data = []

DATADIR = '/content/images'
CATEGORIES = ['Classic Bicycles', 'building', 'soccer ball']

for category in CATEGORIES:
  class_num = CATEGORIES.index(category)
  path = os.path.join(DATADIR,category)  #đường dẫn tới từng mục /content/images/...
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))  # value từng bức ảnh trong 3 mục
    #print(img_array.shape)  #(1200, 1600, 3) (height, width, depth)
    #plt.imshow(img_array)
    img_resized = resize(img_array,(150,150,3))   #quy về giá trị từ 0 đến 1 (pixel có giá tri tù 0 đến 255)
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)
  
flat_data = np.array(flat_data)
images = np.array(images)
target = np.array(target)
```

```python
#Split dataset into training set and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
parameter_grid = [
        {'C':[1,10,100,1000],'kernel':['linear']},
        {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,parameter_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print("{:.2f}%".format(accuracy_score(y_pred, y_test)*100))
 #in ra xác suất đúng
```

```python
confusion_matrix(y_pred, y_test)
```

```python
#Save the model using Pickle library
import pickle
pickle.dump(clf, open('img_model.p','wb'))   #writing byte "img_model.p"
```
